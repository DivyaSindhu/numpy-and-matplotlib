```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
days=[1,2,3,4,5,6,7]
temp=[50,51,52,48,47,46,49]
min_temp=[30,32,28,29,30,32,31]
max_temp=[52,53,54,50,49,48,50]
```

```python

plt.plot(days,temp,label="avg")
plt.plot(days,min_temp, label="min")
plt.plot(days,max_temp, label="max")

plt.legend()

plt.grid()
```
